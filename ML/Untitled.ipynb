```python
import random
import pandas as pd
import numpy as np
```

```python
distance=[]
for i in range(1000):
    distance.append(random.randint(0,10000))
```

```python
engineOil=[]
for i in range(1000):
    engineOil.append(random.randint(0,10))
```

```python
tireAge=[]
for i in range(1000):
    tireAge.append(random.randint(0,10))
```

```python
emissions=[]
for i in range(1000):
    emissions.append(random.randint(0,100))
```

```python
batteryAge=[]
for i in range(1000):
    batteryAge.append(random.randint(0,10))
```

```python
score=[]
for i in range(1000):
    score.append(random.randint(0,10))
```

```python
data={
    "distance":distance,
    "engineOil":engineOil,
    "tireAge":tireAge,
    "emissions":emissions,
    "batteryAge":batteryAge,
    "score":score
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
X=df.drop('score',axis=1)
y=df['score']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
scaled=scaler.fit_transform(X)
X=scaled
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
import pickle
```

```python
pkl_filename = "pickle_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)

# Load from file
with open(pkl_filename, 'rb') as file:
    pickle_model = pickle.load(file)
    
# Calculate the accuracy score and predict target values
score = pickle_model.score(X, y)
print("Test score: {0:.2f} %".format(100 * score))
Ypredict = pickle_model.predict(X)
```
